# Exploratory Data Analysis — Moroccan Gold Prices

Monthly time series (2000-08 → 2026-02). Target: **gold_price_mad** = USD gold × USD/MAD.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().resolve().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import matplotlib.pyplot as plt

from src.evaluation.diagnostics import run_eda_diagnostics
from src.preprocessing.data_loader import load_and_prepare_data
from src.utils.config import EVENT_COLUMNS, FIGURES_DIR, TARGET_COL

df = load_and_prepare_data()
print(df.shape, df.index.min(), "→", df.index.max())
df.head()
```

```python
df[TARGET_COL].plot(figsize=(12, 4), title="Moroccan gold price (MAD)")
plt.ylabel("MAD")
plt.show()

run_eda_diagnostics(df)
print("Figures saved to", FIGURES_DIR)
```

```python
df[EVENT_COLUMNS].plot(figsize=(12, 5), subplots=True, title="Event intensity")
plt.tight_layout()
plt.show()

df[EVENT_COLUMNS].corrwith(df[TARGET_COL]).sort_values(ascending=False)
```
